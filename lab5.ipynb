```python
import seaborn as sns
import matplotlib.pyplot as plt
df_tips = sns.load_dataset('tips')
df_flights= sns.load_dataset('flights')
df_titanic = sns.load_dataset('titanic')
df_iris = sns.load_dataset('iris')
df_diamonds = sns.load_dataset("diamonds")
```

```python
##1. 
sns.regplot(data=df_tips, x="total_bill", y="tip",
            x_jitter=.75, 
            ci=.1, 
            marker="+",
            line_kws=dict(color='g'))
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.title("Regression Graph")
sns.set_theme(style="darkgrid")
```

```python
##2.
sns.lmplot(data=df_titanic, 
            x="age", 
            y="fare", 
            hue="class", 
            col="sex", 
            height=5,   
            aspect=1.2,  
            scatter_kws={'s': 50, 'alpha': 0.5}, 
            markers=["o", "s", "D"], 
            line_kws={'linewidth': 2})

plt.show()
```

```python
##3.
flights_heat_map = df_flights.pivot(index="year", columns="month", values="passengers")
plt.figure(figsize=(12, 8))
hm = sns.heatmap(flights_heat_map, 
            annot=True, # to have values on squares 
            linewidths=.5,
            fmt=".1f", #format
            cmap="magma", # colortheme
            cbar=True)

plt.xticks(rotation=45) # rotating for better readability
plt.yticks(rotation=0)

plt.title("Heatmap of passengers in flight dataset")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()



```

```python
##4.
species = df_iris.pop("species")
sns.clustermap(df_iris,
               method="ward",
               metric="euclidean",
               row_cluster=False,
               cmap="coolwarm")
```

```python
##5.
df_iris1 = sns.load_dataset("iris")
sns.pairplot(df_iris1, 
             hue="species", 
             markers=['o','s','D'], 
             plot_kws={'alpha':0.7})
```

```python
##6.
plt.figure(figsize=(12,8))
sns.jointplot(df_tips,
              x="total_bill",
              y="tip",
              kind="hex",
              marginal_kws=dict(bins=25))
```

```python
##7.
g = sns.FacetGrid(
    df_titanic, 
    col="class", 
    row="sex", 
    #col_wrap=3, this gives error but no idea how to fix this
    height=4, 
    aspect=1.2, 
    despine=False 
)

g.map(sns.histplot, "age", bins=20, kde=True)
g.set_titles(col_template="{col_name} Class", row_template="{row_name}")
g.set_axis_labels("Age", "Count")

plt.show()
```

```python
##8.
plt.figure(figsize=(12, 6))
kinds = ["bar", "box", "strip", "violin"]

for kind in kinds:
    g = sns.catplot(
        data=df_tips, 
        x="day", 
        y="total_bill", 
        kind=kind, 
        hue="sex", 
        dodge=True,
        height=5, 
        aspect=1.5
    )
    
    g.set_xticklabels(rotation=45)
    g.figure.suptitle(f"Categorical Plot ({kind})", fontsize=14)

    plt.show()


```

```python
##9.
plt.figure(figsize=(12,12))
g = sns.lmplot(
    data=df_diamonds, 
    x="carat", 
    y="price", 
    hue="cut", 
    col="cut",
    palette="muted",
    ci=50,
    height=4,  
    aspect=1
)
g.figure.suptitle("Regression Plot of Carat vs Price (Faceted by Cut)", fontsize=10)

plt.show()
```
