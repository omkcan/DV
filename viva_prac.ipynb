```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Problem Statement - 

```python
tips = sns.load_dataset("tips")

#bar plot
ax = sns.barplot(data=tips, x="day", y="total_bill", hue="time")
for bar in ax.patches:
    bar.set_width(0.5)
plt.grid(True, linestyle="--")
plt.xlabel("Days")
plt.ylabel("Total Bill")
plt.xticks(rotation=45)
plt.show() 
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",alpha=0.8)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="tips",hue="time")
```
