```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('datasets/customer_churn.csv')
```

```python
plt.rcParams['figure.figsize'] = [5,5]

plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Bar graph sir')
plt.bar(data['Gender'],data['Age'])

```

Histogram

```python
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.hist(data['Age'])
plt.show()
```

Pie chart

```python
labels = 'IT', 'Marketing', 'Data Science', 'Finance'
values = [500, 150, 300, 510]
explode = (0.05, 0.05, 0.05, 0.05)

#plt.pie(values, labels=labels)
plt.pie(values, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
df = pd.read_csv('datasets/titanic_data.csv')

labels = '1st Class', '2nd Class', '3rd Class'
values = df['Pclass'].value_counts()
explode = (0.05, 0.05, 0.05)

plt.pie(values, labels=labels, autopct='%1.1f%%', shadow=True, explode=explode)
plt.show()
```

Stack Area Chart

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
London = [25, 26, 32, 19, 28, 39, 24]
Tokyo = [20, 29, 23, 35, 32, 26, 18]
Paris = [18, 21, 28, 35, 29, 25, 22]

plt.figure(figsize=(10, 6))
plt.stackplot(days, London, Tokyo, Paris, labels=["London", "Tokyo", "Paris"], colors=["green", "white", "orange"], alpha=0.8)
plt.xlabel("Days of the Week", fontsize=12)
plt.ylabel("Temperature (C)", fontsize=12)
plt.title('Days of Week v/s Temperature')
plt.legend(loc="upper right")
plt.show()
```
