```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

Q1

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", markers=["o", "s"], col="time")
plt.show()
```

Q2

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
flights_corr = flights_pivot.corr()
sns.heatmap(flights_corr, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Correlation Matrix of Flights Data")
plt.show()
```

Q3

```python
sns.clustermap(titanic[["pclass", "age"]].dropna(), method="ward", cmap="viridis", standard_scale=1)
plt.show()
```

Q4

```python
sns.pairplot(iris, hue="species", kind="reg", palette="Set2")
plt.show()
```

Q5

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex", alpha=0.5)
plt.show()
```

Q6

```python
sns.catplot(data=titanic, x="pclass", y="fare", kind="box", hue="sex")
sns.swarmplot(data=titanic, x="pclass", y="fare", hue="sex", dodge=True)
plt.xticks(rotation=45)
plt.show()
```

Q7

```python

sns.relplot(data=titanic, x="age", y="survived", hue="survived", size="survived", alpha=0.5)
plt.legend(loc="best")
plt.show()
```

Q8

```python

sns.lmplot(data=iris, x="petal_length", y="petal_width", hue="species", aspect=1.5, height=6, markers=["o", "s", "D"])
plt.show()
```

Q9

```python
fg = sns.FacetGrid(flights, col="year", col_wrap=4)
fg.map(sns.lineplot, "month", "passengers", linestyle="dashed")
plt.show()
```
