```python
import numpy as np
import matplotlib.pyplot as plt

#idk what this does
plt.rcParams["figure.figsize"] = [5,5]

#this I know
x_val = np.linspace(0,5,20)
y_val = x_val**2
y_val2 = x_val

#as usual plotting
plt.plot(x_val, y_val,'r',label='y=x^2')
plt.plot(x_val,y_val2,'b',label='y=x')
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('a simple graph')
plt.legend(loc='best') 
```
