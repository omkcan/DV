```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris_data = sns.load_dataset('iris')
```

```python
x_vals = np.linspace(0, 20, 20)
y_vals = 2*x_vals + 5
y2_vals = x_vals ** 3
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Linear Relation')
plt.plot(x_vals, y_vals, 'r', label = 'Linear Relation')
plt.plot(x_vals, y2_vals, 'b', label = 'Cube')
plt.legend(loc='best') #location mate loc use karvanu 
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal length vs Sepal width')
plt.scatter(iris_data['sepal_length'], iris_data['petal_length'], c = 'r', label = 'Petal Length', marker='+')
plt.scatter(iris_data['sepal_length'], iris_data['sepal_width'], c = 'b', label = 'Sepal Width', marker='s')
plt.legend(loc='best') 
```
