1.Write code for line chart to generate 70 numbers with X- axis and Y-axis is
mx+c, were m=2 and c=0.3, Draw the four different line chart using matplotlib
library. (Take Y1=X, Y2=X2, Y3=X3 and Y4=sqrt(X)).

```python
import numpy as np
import matplotlib.pyplot as plt

m = 2
c = 0.3

X = np.linspace(0, 35, 70)  

Y = m * X + c
Y1 = X
Y2 = X ** 2
Y3 = X ** 3
Y4 = np.sqrt(X)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot Y1 = X
axs[0, 0].plot(Y, Y1, label='Y1 = X', color='b')
axs[0, 0].set_title('Y1 = X')
axs[0, 0].set_xlabel('Y = 2X + 0.3')
axs[0, 0].set_ylabel('Y1 = X')
axs[0, 0].legend()
axs[0, 0].grid()

# Plot Y2 = X^2
axs[0, 1].plot(Y, Y2, label='Y2 = X^2', color='g')
axs[0, 1].set_title('Y2 = X^2')
axs[0, 1].set_xlabel('Y = 2X + 0.3')
axs[0, 1].set_ylabel('Y2 = X^2')
axs[0, 1].legend()
axs[0, 1].grid()

# Plot Y3 = X^3
axs[1, 0].plot(Y, Y3, label='Y3 = X^3', color='r')
axs[1, 0].set_title('Y3 = X^3')
axs[1, 0].set_xlabel('Y = 2X + 0.3')
axs[1, 0].set_ylabel('Y3 = X^3')
axs[1, 0].legend()
axs[1, 0].grid()

# Plot Y4 = sqrt(X)
axs[1, 1].plot(Y, Y4, label='Y4 = sqrt(X)', color='m')
axs[1, 1].set_title('Y4 = sqrt(X)')
axs[1, 1].set_xlabel('Y = 2X + 0.3')
axs[1, 1].set_ylabel('Y4 = sqrt(X)')
axs[1, 1].legend()
axs[1, 1].grid()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

2. Plot the function y = x - 4x + 3 for x ranging from -2 to 6. Identify and
annotata the minumum point of the graph. Add appropriate labels, title and
grid.

```python
# Define X values
X = np.linspace(-2, 6, 100)

# Define the function y = x^2 - 4x + 3
Y = X**2 - 4*X + 3

# Find the minimum point using vertex formula x = -b/(2a)
a, b, c = 1, -4, 3
x_min = -b / (2 * a)
y_min = x_min**2 - 4*x_min + 3

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(X, Y, label='y = x - 4x + 3', color='b')
plt.scatter(x_min, y_min, color='r', marker='o', label='Minimum Point')
plt.annotate(f'Minimum ({x_min:.2f}, {y_min:.2f})', (x_min, y_min), 
             textcoords="offset points", xytext=(-30,-10), ha='center', color='r')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot of y = x - 4x + 3')
plt.legend(loc='best')
plt.grid()

# Show the plot
plt.show()
```

3. Plot sin(x) and cos(x) from 0 to 2 on the same graph. Use fill_between()
to shade the area between the two curves where sin(x) is greater than cos(x).
Add labels and a title.

```python
# Define X values
X = np.linspace(0, 2 * np.pi, 100)

# Define functions
Y_sin = np.sin(X)
Y_cos = np.cos(X)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(X, Y_sin, label='sin(x)', color='b')
plt.plot(X, Y_cos, label='cos(x)', color='r')

# Shade the area where sin(x) > cos(x)
plt.fill_between(X, Y_sin, Y_cos, where=(Y_sin > Y_cos), interpolate=True, color='grey', alpha=0.3)

# Add labels and title
plt.xlabel('X-axis (radians)')
plt.ylabel('Y-axis')
plt.title('Plot of sin(x) and cos(x) with Shaded Region')
plt.legend(loc='best')
plt.grid()

# Show the plot
plt.show()
```

4.Consider monthly temperature data for the year 2023 and plot it using a
time-line graph.

```python
# Define months and temperature data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperature = [12, 14, 18, 24, 30, 34, 32, 31, 28, 22, 16, 13]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(months, temperature, marker='o', linestyle='-', color='b', label='Temperature')

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Temperature (C)')
plt.title('Monthly Temperature Data for 2023')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

5.Plot three mathematical functions on the same graph. Add a legend, labels,
and title.

```python
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = 2*x + 1
y3 = np.sqrt(x)
plt.figure(figsize=(4,3))
plt.plot(x, y1, 'r--', label='y = x^2')
plt.plot(x, y2, 'b-', label='y = 2x + 1')
plt.plot(x, y3, 'g:', label='y = sqrt(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Comparison of Functions')
plt.legend()
plt.grid(True)
plt.show()
```

6. Create a dual-axis plot where:- The primary y-axis represents the monthly rainfall
(in mm) in 2023 as a bar chart.- The secondary y-axis represents the average tem-
perature (in C) as a line graph on the same plot.

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
rainfall = [78, 62, 55, 41, 32, 140, 300, 280, 150, 90, 85, 80] # in mm
temperature = [12, 14, 18, 24, 30, 34, 32, 31, 28, 22, 16, 13] # in C
fig, ax1 = plt.subplots(figsize=(5,3))
ax1.bar(months, rainfall, color='skyblue', label='Rainfall (mm)')
ax1.set_xlabel('Month')
ax1.set_ylabel('Rainfall (mm)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax2 = ax1.twinx()
ax2.plot(months, temperature, color='orange', marker='o', label='Temperature(C)')
ax2.set_ylabel('Temperature (C)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')
plt.title('Monthly Rainfall and Average Temperature in 2023')
fig.tight_layout()
```

7.Draw a house using matplotlib line charts

```python
fig, ax = plt.subplots(figsize=(6, 6))

walls_x = [1, 1, 5, 5, 1]
walls_y = [1, 4, 4, 1, 1]

ax.plot(walls_x, walls_y, color='brown', label="Walls")

roof_x = [1, 3, 5]
roof_y = [4, 6, 4]

ax.plot(roof_x, roof_y, color='red', label="Roof")

door_x = [2, 2, 3, 3, 2]
door_y = [1, 3, 3, 1, 1]

ax.plot(door_x, door_y, color='blue', label="Door")

window1_x = [4, 4, 4.5, 4.5, 4]
window1_y = [3, 3.5, 3.5, 3, 3]

ax.plot(window1_x, window1_y, color='teal', label="Window 1")

window2_x = [4, 4, 4.5, 4.5, 4]
window2_y = [1.5, 2, 2, 1.5, 1.5]

ax.plot(window2_x, window2_y, color='purple', label="Window 2")
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_title("House Diagram", fontsize=14)
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)
ax.legend()
plt.show()
```
