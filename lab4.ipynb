```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
df2 = sns.load_dataset('flights')
df3 = sns.load_dataset('titanic')
```

1.Create a scatter plot to visualize the relationship between total_bill and tip in the tips dataset.

```python
sns.scatterplot(data = df ,x='total_bill',y='tip',hue='time',style='sex',size='size')
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.title("Scatter Plot")
plt.show()
```

This scatter plot visualizes the relationship between total bill and tips in the **tips** dataset. Different colors represent meal time (Lunch/Dinner), while markers differentiate sex and marker sizes indicate party size.

2.Use the flights dataset to create a line plot showing the trend of passengers over months.

```python
sns.lineplot(data=df2,x='month',y='passengers',hue='year',style='year',errorbar='sd')
plt.xlabel('Months')
plt.ylabel('Passengers')
plt.title("Line Graph")
plt.legend(loc='best')
plt.show()
```

From this line graph , we can observe a surge in the no of passengers between the months of June and September, which then slowly decreases.

3.Visualize the relationship between total_bill and tip in the tips dataset using a relational plot.

```python
sns.relplot(data=df,x='total_bill',y='tip',col='smoker',row='sex',markers=['*','+','x','s'],height=5,aspect=1.5)
plt.title("Relational Plots")
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.show()
```

This relational plot creates a grid of scatter plots, splitting the data by **smoker** (columns) and **sex** (rows) from the **tips** dataset. Different markers are used to differentiate groups, showing the relationship between **total bill** and **tip** amounts.

4.Create a bar plot to compare the average total_bill for each day in the tips dataset.

```python
#4.
sns.barplot(data=df,x='day',y='total_bill',hue='sex',errorbar='sd')
plt.xlabel('Days')
plt.ylabel('Total Bill')
plt.title("Bar Chart")
plt.show()
```

From this bar chart , we can observe that the total bill of both Males and Females is significantly higher in Saturdays and Sundays.

5.Use a count plot to show the distribution of passengers by their class in the titanic dataset.

```python
#5.
sns.countplot(data=df3,x='pclass',orient='h',order=[1,2,3])
plt.xlabel('PClass')
plt.title("Count Plot")
plt.show()
```

 From the count plot , we can see that the count of PClass attribute 3 is more than double when compared to the other values of PClass. 

6.Plot the distribution of age for each class in the titanic dataset using a box plot.

```python
sns.boxplot(data=df3,x='pclass',y='age',hue='sex',notch=True)
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title("Box Plot")
plt.show()
```

This **box plot** displays the **age distribution** across different **passenger classes (Pclass)** in the **Titanic** dataset, categorized by **sex**. The **notches** indicate the confidence intervals for the median age in each group.

7.Compare the fare distribution for passengers based on their embarked location in the titanic dataset using a violin plot.

```python
sns.violinplot(data=df3,x='embarked',y='fare',hue='sex',inner='quart',split=True)
plt.xlabel('Embarked')
plt.ylabel('Fare')
plt.title("Violin Plot")
plt.show()
```

This **violin plot** visualizes the **fare distribution** for passengers based on their **embarkation point** in the **Titanic** dataset, separated by **sex**. It combines aspects of a **box plot** and **KDE (Kernel Density Estimation)** to show data spread and quartiles.

8.Create a strip plot to display the distribution of tip for different days in the tips dataset.

```python
sns.stripplot(data=df,x='day',y='tip',hue='sex',jitter=True,alpha=0.7)
plt.xlabel('Days')
plt.ylabel('Tips')
plt.title("Stripper Plot")
plt.show()
```

From this stripper plot, we see that the tips on Saturday are usually of a larger value as compared to tips on other days.

9.Use a swarm plot to visualize the distribution of tip for different days in the tips dataset.

```python
sns.swarmplot(data=df,x='day',y='tip',hue='sex',size=7)
plt.xlabel('Days')
plt.ylabel('Tips')
plt.title("Swarm Plot")
plt.show()
```

Here, we see that the tips on Saturday are usually of a larger value as compared to tips on other days.

10.Plot the distribution of fare in the titanic dataset using a histogram.

```python
sns.histplot(data=df3['fare'],kde=True,bins=30,edgecolor='black')
plt.title("Hist Plot")
plt.show()
```

From the histplot ,we can see that the frequency of the fare price between 0 to 50 is significantly larger than all other fare prices. 

11.Use a KDE plot to compare the distributions of total_bill for sex in the tips dataset.

```python
sns.kdeplot(data=df, x='total_bill', hue='sex', fill=True, common_norm=False)
plt.xlabel('Total Bill')
plt.title("KDE Plot")
plt.show()
```

The KDE plot shows a bell shaped or normal distribution of total_bill , with a median value between 10 and 30.

12.Compare the total_bill distribution for different sex categories in the tips dataset using a histogram.

```python
sns.histplot(data=df, x='total_bill', hue='sex', kde=True, element='step', common_norm=False)
plt.xlabel('Total Bill')
plt.title("Hist Plot")
plt.show()
```

Here, the Histplot shows that the frequency of total bill between 10 and 30 is much much more than any other amount of total bill.
