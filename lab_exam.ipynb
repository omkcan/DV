Titanic Dataset (Passenger Survival Analysis)

```python
#Global Imports
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")
```

```python
#Preview of the dataset
titanic
```

Q1. Create a Barplot

```python
plt.figure(figsize=(10, 6))
g = sns.countplot(y='pclass', hue='survived', data=titanic, palette=['red', 'green'])

plt.title("Survival Count by Passenger Class")
plt.ylabel("Passenger Class")
plt.xlabel("Count")         
plt.yticks(ticks=[0, 1, 2], labels=["1st Class", "2nd Class", "3rd Class"],rotation=45)
plt.legend(title="Survival Status", labels=["Not Survived", "Survived"])

for p in g.patches:
    width = p.get_width()  
    if width > 0:         
        g.annotate(f'{int(width)}', 
                    (width, p.get_y() + p.get_height() / 2.),  
                    ha='left', va='center',                
                    xytext=(5, 0),
                    textcoords='offset points')

plt.show()
```

Q2. Boxplot

```python
plt.figure(figsize=(10,7))
sns.swarmplot(data=titanic,x="pclass",y="age",color="blue", alpha=0.5, size=6)
sns.boxplot(data=titanic,x="pclass",y="age",palette="coolwarm")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Age Distribution by Passenger Class")
plt.xticks(rotation=30)
plt.show()
```

Q3. Violin Plot

```python
plt.figure(figsize=(10,7))
sns.violinplot(data=titanic,y="fare",hue="sex",split=True,alpha=0.9,inner="quartile",palette=["blue","pink"])
plt.xlabel("Sex")
plt.ylabel("Fare")
plt.title("Fare Distribution by Sex")
plt.show()
```
