```python
from scipy.stats import linregress

budget = np.random.randint(0,201,60)
sales_revenue = np.random.randint(0,180,60)

plt.scatter(budget, sales_revenue, marker="o", color="blue", label="Data points")

slope, intercept, _, _, _ =linregress(budget, sales_revenue)
line = slope*budget + intercept
plt.plot(budget, line, color='red', label="Reg line")

plt.xlabel("Budget")
plt.ylabel("Sales Revenue")
plt.title("Scatter Plot with regression line")
plt.show()
```

4. The purpose of the update_layout() method in Plotly is to customize the layout and appearance of a chart  such as titles, axis labels, legends, margins, colors, and interactivity.

The key difference between a bar chart and a histogram in Plotly lies in what they represent and how the data is grouped:

 Bar Chart (Categorical Data)
Used for: Comparing values across distinct categories.

X-axis: Categorical labels (e.g., "Apple", "Banana", "Mango").

Bars have gaps between them.

Built with: go.Bar() or px.bar().

 Histogram (Continuous Data)
Used for: Showing the distribution of numerical (continuous) data.

X-axis: Intervals or bins (e.g., age ranges like 010, 1020).

Bars are touching to show continuity.

Built with: px.histogram() or go.Histogram().

2. Sales Performance Report
Scenario: A dataset (sales.csv) contains sales by product and region.
Task:
 Plot a stacked bar chart comparing sales across regions.
 Add a button to toggle between stacked and grouped views.
 Highlight the best-selling product in each region.
 Add a title and axis labels.

```python
df = pd.read_csv("datasets/sales.csv")
regions = df["Region"].unique()
products = df["Product"].unique()

# Prepare traces
stacked_traces = [
    go.Bar(name=prod,
           x=regions,
           y=df[df["Product"] == prod].groupby("Region")["Sales"].sum())
    for prod in products
]
grouped_traces = [
    go.Bar(name=prod,
           x=regions,
           y=df[df["Product"] == prod].groupby("Region")["Sales"].sum())
    for prod in products
]

# Highlight best seller
best = df.groupby(["Region", "Product"])["Sales"].sum().reset_index()
best = best.loc[best.groupby("Region")["Sales"].idxmax()]

# Build figure
fig = go.Figure(data=stacked_traces)

# Toggle button
fig.update_layout(
    barmode="stack",
    title="Sales by Product and Region",
    xaxis_title="Region",
    yaxis_title="Sales",
    updatemenus=[{
        "buttons": [
            {"label": "Stacked",
"method": "update",
             "args": [{"visible": [True]*len(products)},
                      {"barmode": "stack"}]},
            {"label": "Grouped",
             "method": "update",
             "args": [{"visible": [True]*len(products)},
                      {"barmode": "group"}]},
        ],
        "direction": "down", "showactive": True, "x": 0.0, "y": 1.2
    }]
)

# Add annotations for best-sellers
for _, row in best.iterrows():
    fig.add_annotation(
        x=row["Region"],
        y=row["Sales"] + df["Sales"].max()*0.05,
        text=f"Top: {row['Product']}",
        showarrow=False,
        font=dict(color="black", size=12)
    )

fig.show()
```

3.COVID-19 Data Visualization
Scenario: A dataset (covid.csv) has daily cases and deaths for 5 countries.

```python


# Load the data
df = pd.read_csv("datasets/covid.csv", parse_dates=["Date"])

# Add a string date column for animation frame
df["date_str"] = df["Date"].dt.strftime("%Y-%m-%d")

# Bubble chart
fig = px.scatter(
    df, 
    x="Cases", 
    y="Deaths", 
    size="Recoveries", 
    animation_frame="date_str",  # Use the string column instead of the computed value
    color="Country", 
    hover_name="Country",
    hover_data={"Date": True, "Cases": True, "Deaths": True},
    title="COVID-19: Cases vs Deaths (Bubble Size = Recoveries)"
)

# Fix update_layout to enable actual PNG download button
fig.update_layout(
    updatemenus=[{
        "type": "buttons",
        "buttons": [
            dict(
                args=["toImageButtonOptions", {"format": "png", "filename": "covid_chart"}],
                label="Download PNG",
                method="relayout"
            )
        ]
    }]
)

fig.show()
```

4. Iris Dataset Analysis
Scenario: Use the Iris dataset (iris.csv) with sepal/petal dimensions.
Task:
 Generate a violin plot comparing sepal length by species.
 Overlay a box plot on the same figure.
 Add a legend toggle to hide/show species.
 Label axes and add a title.

```python
import plotly.express as px

iris_data = px.data.iris()

fig = px.violin(
    iris_data,
    x="species",
    y="sepal_length",
    color="species",
    box=True,
    points="all",
    title="Sepal lenght by Species",
    labels={
        "species":"Species",
        "sepal_length":"Sepal Length (cm)"
    }
)

fig.update_layout(
    updatemenus=[
        dict(
           type="buttons",
            direction="left",
            buttons=[
                dict(
                    args=[{"visible": [True, True, True]}, {"showlegend": True}],
                    label="Show Legend",
                    method="update"
                ),
                dict(
                    args=[{"visible": [True, True, True]}, {"showlegend": False}],
                    label="Hide Legend",
                    method="update"
                )
            ],
        )
    ]
)

fig.show()
```
